Create all combinational datasets

```python
def create_comb_data(df1,df2):
    df = pd.concat([df1,df2],axis=0)
    #Shuffle the dataset
    df = df.sample(frac=1)
    df = df.reset_index()
    df.drop('index',axis=1,inplace=True)
    print('Shape of dataset: ',df.shape)
    return df
```

### 3.1 Split the dataset in Train & Test

```python
def split_train_test(df):
    X = df.drop(['assetid','level'],axis=1)
    y = df['level']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify= y,random_state=100)
    return X_train, X_test, y_train, y_test
```

```python
data_review = pd.read_csv('/Users/vikaschaturvedi/Documents/T-Systems/POC/Recommender/Yelp-Recommender-master/data/interim/reviews.csv')
print(data_review.shape)
data_review.head()
```

>Let's get all unique business ids

```python
business_id_ls = data_business.business_id.unique()
```

```python
# keep only b_id in b_id ls
review_idx = [i for i in range(data_review.shape[0]) if data_review.business_id[i] in business_id_ls]

len(review_idx)
```

```python
review_df = data_review.iloc[review_idx,:]
```

```python
review_df = review_df.drop(columns=['Unnamed: 0'])
```

```python
review_df = data_review.merge(data_business[['business_id']],on=['business_id'],how='inner')
print(review_df.shape)
review_df.head()
```

```python
print('Unique Business Ids:',review_df.business_id.nunique())
print('Unique Review Ids:',review_df.review_id.nunique())
```

```python
review_df.index = range(review_df.shape[0])
```

```python
yr_ls = [eval(x.split('-')[0]) for x in review_df.date]
review_df['year'] = yr_ls
review_df.head()
```

```python
# write csv
review_df.to_csv('/Users/vikaschaturvedi/Documents/T-Systems/POC/Recommender/Yelp-Recommender-master/data/interim/reviews_atlanta.csv',index=False)
```

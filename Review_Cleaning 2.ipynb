```python
data_review.drop(['stars','average_stars','stars_x'],axis=1,inplace=True)
```

```python
data_business = pd.read_csv('/Users/vikaschaturvedi/Documents/T-Systems/POC/Recommender/Yelp-Recommender-master/data/interim/atlanta_business_Nora2.csv')
print(data_business.shape)
data_business.head()
```

```python
data_users = pd.read_csv('/Users/vikaschaturvedi/Documents/T-Systems/POC/Recommender/Yelp-Recommender-master/data/interim/users_atlanta.csv')
print(data_users.shape)
data_users.head()
```

```python
user_tag = data_review.merge(data_business, left_on='business_id', right_on='business_id')
print(user_tag.shape)
user_tag.head()
```

```python
user_tag.columns
```

```python
user_tag.iloc[:,33:].head()
```

```python
user_tag2 = user_tag.groupby('user_id').sum().iloc[:,30:]
user_tag2.shape
```

```python
user_tag2.head()
```

```python
tag = []
for i in range(user_tag2.shape[0]):
    tmp = user_tag2.iloc[i,:].sort_values(ascending=False)[:2].index.tolist()
    tag.extend(tmp)

len(set(tag))
```

```python
tag_all = user_tag2.idxmax(axis=1)
tag_list = list(set(tag_all))
tag_all_df = pd.DataFrame(tag_all).reset_index()

for tag in tag_list:
    data_users[tag] = 0
```

```python
for ind,val in enumerate(tag_all):
    row = list(data_users['user_id']).index(tag_all_df['user_id'][ind])
    data_users.loc[row,val] = 1
```

```python
data_review = pd.read_csv('/Users/vikaschaturvedi/Documents/T-Systems/POC/Recommender/Yelp-Recommender-master/data/interim/reviews_atlanta.csv')
print(data_review.shape)
data_review.head()
```

```python
print('Total Unique Users in Atlanta:', data_review.user_id.nunique())
print('Total Unique Businesses in Atlanta:', data_review.business_id.nunique())
print('Total Unique Reviews given by those users for the Businesses in Atlanta:', data_review.review_id.nunique())
```

```python
unique_atlanta_users = data_review.user_id.unique()
df_users = pd.DataFrame(unique_atlanta_users,columns=['user_id'])
print(df_users.shape)
df_users.head()
```

```python
user_df = data_user.merge(df_users,on='user_id',how='inner')
print(user_df.shape)
user_df.head()
```

```python
# fix indexes
user_df.index = range(user_df.shape[0])
```

```python
# check na
user_df.isnull().sum().sort_values(ascending=False) # elite has many NA, better make binary
```

```python
# feature engineering
user_df['is_elite'] = ~np.array(user_df.elite.isnull())
```

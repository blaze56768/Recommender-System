Create all combinational datasets

```python
def create_comb_data(df1,df2):
    df = pd.concat([df1,df2],axis=0)
    #Shuffle the dataset
    df = df.sample(frac=1)
    df = df.reset_index()
    df.drop('index',axis=1,inplace=True)
    print('Shape of dataset: ',df.shape)
    return df
```

### Feature Selection

```python
def correlation(df,threshold):

    # Create correlation matrix
    corr_matrix = df.corr().abs()

    # Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

    # Find features with correlation greater than threshold
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    print("Number of columns to drop : ",len(to_drop))
    print("Proportions of columns to drop : ",round(len(to_drop)/len(df.columns)*100,2),"%")
    # Drop features 
    df1 = df.drop(to_drop, axis=1)
    
    return df1
```

### 3.1 Split the dataset in Train & Test

```python
def split_train_test(df):
    X = df.drop(['assetid','level'],axis=1)
    y = df['level']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify= y,random_state=100)
    return X_train, X_test, y_train, y_test
```
